# Type

```python
dt.type.unique()
```

```python
dt.type.value_counts()
```

```python
# Finding the mean for the average price by grouping the type
dt.groupby('type').AveragePrice.mean().reset_index()
```

```python
a=sns.catplot(x='type',y='AveragePrice',data=dt,palette='husl',height=10,kind='boxen')
plt.title('Boxen plot of AverigePrive for each type',size=25)
```

from above plot the obsevations are:

     1) conventional type have low avearge price. i.e is 04 to 2.3 (approximately)
     2) organic type have the high average price. i.e is 0.4 to 3.25 (approximately)

```python
a=sns.catplot(x='type',y='Total Volume',data=dt,palette='husl',height=10,kind='boxen')
plt.title('Boxen plot of Total Volume for each type',size=25)
```

from above plot the obsevations are:

    1) the coventional type has high total number of avocados sold.
    2) the organic type has low total number of avocados sold compare to convention type

```python
sns.relplot(x="Total Volume",y='AveragePrice',hue='type',data=dt,height=10)
plt.title('AveragePrice Vs Total Volume for each type',size=25)
```

from above plot the obsevations are:
    
   1) We can show a relation between the Average Price and Total Volume.
   
   2) we can see that the organic type has high Average Price and low total volume.
   
   3) we can see that the conventional type has low Average Price and high total volume

```python
plt.figure(figsize=(13,15))
a=sns.barplot(x='AveragePrice',y='region',data=dt,palette='husl',hue='type')
a.set_yticklabels(a.get_yticklabels(),size=16)
plt.title('Barplot AveragePrice Vs Region for each Type',size=30)
```

from above plot the obsevations are:

    1) in every region the coventional type has low average price and organic region has high average price.

```python
plt.figure(figsize=(10,15))
a=sns.barplot(x='Total Volume',y='region',data=dt[dt.type=='organic'].query('region != "TotalUS"'),palette='coolwarm')
a.set_yticklabels(a.get_yticklabels(),size=16)
plt.title('Total Volume for organic for each Region',size=30)
```

from above plot the obsevations are:
    1) West region has high total volume of organic type ,exceeding the totalus

```python
plt.figure(figsize=(10,15))
a=sns.barplot(x='Total Volume',y='region',data=dt[dt.type=='conventional'].query('region != "TotalUS"'),palette='coolwarm')
a.set_yticklabels(a.get_yticklabels(),size=16)
plt.title('Total Volume for conventional for each Region',size=30)
```

from above plot the obsevations are: 1) West region has high total volume of coventional type,exceeding the totalus

# Year

```python
dt.year.unique()
```

```python
dt.Date=pd.to_datetime(dt.Date)
```

```python
dt['month']=dt.Date.dt.month
```

```python
price_years=dt.groupby(['year','month','type'],as_index=False)['AveragePrice'].mean()
price_years
```

Let's see the mean of AveragePrice each month over years

2015

```python
plt.figure(figsize=(13,9))
a=sns.lineplot(x='month',y='AveragePrice',data=price_years.query('year=="2015"')
            ,hue='type',markers=True,style='type'
            ,palette='gnuplot2' )
```

2016

```python
plt.figure(figsize=(13,9))
a=sns.lineplot(x='month',y='AveragePrice',data=price_years.query('year=="2016"')
            ,hue='type',markers=True,style='type'
            ,palette='gnuplot2' )
```

2017

```python
plt.figure(figsize=(13,9))
a=sns.lineplot(x='month',y='AveragePrice',data=price_years.query('year=="2017"')
            ,hue='type',markers=True,style='type'
            ,palette='gnuplot2' )
```

2018

```python
plt.figure(figsize=(13,9))
a=sns.lineplot(x='month',y='AveragePrice',data=price_years.query('year=="2018"')
            ,hue='type',markers=True,style='type'
            ,palette='gnuplot2' )
```

-In 2015, the average price of conventional avocado was in kind of stabilization and did not exceed 1.15 over all months .
-The year 2017 had the highest average price of avocado exactly on september .

# Averge Price of Avocado for each Region over the four years

```python
sns.factorplot('AveragePrice','region',data=dt.query("type=='conventional'"),
                hue='year',
                size=15,
                palette='tab20',
                join=False,
                aspect=0.7,
              )
plt.title('For Conventional',size=25)
```

```python
sns.factorplot('AveragePrice','region',data=dt.query("type=='organic'"),
                hue='year',
                size=15,
                palette='tab20',
                join=False,
                aspect=0.7,
              )
plt.title('For Organic',size=25)
```

Results :
the organic avocado was expensive in 2017 especially on sanFrancisco and hartford-springfield
